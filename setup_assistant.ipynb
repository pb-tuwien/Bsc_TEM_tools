# Setup

This notebook should help with the installation of the `Modelling` package.  
**This package needs the Python version 3.11.x to work.**  
We recommend using this package in an `Virtual Environment`. 
The following tutorial also shows how you can create one using [Anaconda](https://www.anaconda.com/).


## Installation of Package

There are to ways of installing this package:

- Through the ``Terminal`` 
- By running this Notebook. This does not always work.

### Through the Terminal

First navigate to the folder with this python package:
```bash
cd your/path/to/Modelling
```
Now your Terminal should show something like `...\Modelling>`.

#### Virtual Environment

##### New Environment
If you want to create a new environment from scratch run:
```bash
conda env create -f tem_forward.yml
```
This created a new environment called `tem_forward`. You can now activate it with:
```bash
conda activate tem_forward
```  
##### Use existing Environment
Else you can activate your existing environment by running:
```bash
conda activate your_env
```
(replace `your_env` with the name of your environment)

#### Installing the package

Now you can install this package.  
For this we use [pip](https://pypi.org/project/pip/). Run the command:
```bash
pip install .
```
This installs the package and makes sure that all dependecies are installed as well.  
The `.` tells pip to look for a Python package in the current folder (particularly for a ``setup.py``-file).  

You could also add the `-e` flag to install the package in "editable"-mode. 
This makes it possible to adapt parts of the package without having to reinstall it after each change.  
For this run:
```bash
pip install -e .
```
